```python
import matplotlib.pyplot as plt
import numpy as np


def drawneuralnetwork():
    fig, ax = plt.subplots()

    # 输入层
    inputlayerx = np.zeros(5)
    inputlayery = np.arange(5)+0.5
    ax.scatter(inputlayerx, inputlayery, c='b', label='Input Layer')

    # 隐藏层1
    hidden1layerx = np.ones(6)*1
    hidden1layery = np.arange(6)
    ax.scatter(hidden1layerx, hidden1layery, c='g', label='Hidden Layer 1')
    for i in range(5):
        for j in range(6):
            ax.plot([inputlayerx[i], hidden1layerx[j]], [inputlayery[i], hidden1layery[j]], 'k')

    # 隐藏层2
    hidden2layerx = np.ones(6)*2
    hidden2layery = np.arange(6)
    ax.scatter(hidden2layerx, hidden2layery, c='r', label='Hidden Layer 2')
    for i in range(6):
        for j in range(6):
            ax.plot([hidden1layerx[i], hidden2layerx[j]], [hidden1layery[i], hidden2layery[j]], 'k')

    # 隐藏层3
    hidden3layerx = np.ones(6)*3
    hidden3layery = np.arange(6)
    ax.scatter(hidden3layerx, hidden3layery, c='y', label='Hidden Layer 3')
    for i in range(6):
        for j in range(6):
            ax.plot([hidden2layerx[i], hidden3layerx[j]], [hidden2layery[i], hidden3layery[j]], 'k')

    # 输出层
    outputlayerx = np.ones(2)*4
    outputlayery = np.arange(2)+2
    ax.scatter(outputlayerx, outputlayery, c='m', label='Output Layer')
    for i in range(6):
        for j in range(2):
            ax.plot([hidden3layerx[i], outputlayerx[j]], [hidden3layery[i], outputlayery[j]], 'k')

    ax.set_xlim(-0.5, 6)
    # ax.set_ylim(-0.5, 6)
    ax.set_title('Deep Neural Network Structure')
    ax.legend(prop={'size': 8})
    plt.show()


drawneuralnetwork()
```

```python
import graphviz


def draw_neural_network():
    dot = graphviz.Digraph(comment='Neural Network')

    # 输入层
    with dot.subgraph(name='cluster_input') as c:
        c.attr(style='filled', color='lightgray')
        c.node_attr.update(style='filled', color='white')
        for i in range(5):
            c.node(f'input_{i}', label=f'x{i}')
        c.attr(label='Input Layer')

    # 隐藏层1
    with dot.subgraph(name='cluster_hidden1') as c:
        c.attr(style='filled', color='lightblue')
        c.node_attr.update(style='filled', color='white')
        for i in range(6):
            c.node(f'hidden1_{i}', label=f'h1_{i}')
        c.attr(label='Hidden Layer 1')
        for i in range(5):
            for j in range(6):
                dot.edge(f'input_{i}', f'hidden1_{j}')

    # 隐藏层2
    with dot.subgraph(name='cluster_hidden2') as c:
        c.attr(style='filled', color='lightgreen')
        c.node_attr.update(style='filled', color='white')
        for i in range(6):
            c.node(f'hidden2_{i}', label=f'h2_{i}')
        c.attr(label='Hidden Layer 2')
        for i in range(6):
            for j in range(6):
                dot.edge(f'hidden1_{i}', f'hidden2_{j}')

    # 隐藏层3
    with dot.subgraph(name='cluster_hidden3') as c:
        c.attr(style='filled', color='lightyellow')
        c.node_attr.update(style='filled', color='white')
        for i in range(6):
            c.node(f'hidden3_{i}', label=f'h3_{i}')
        c.attr(label='Hidden Layer 3')
        for i in range(6):
            for j in range(6):
                dot.edge(f'hidden2_{i}', f'hidden3_{j}')

    # 输出层
    with dot.subgraph(name='cluster_output') as c:
        c.attr(style='filled', color='pink')
        c.node_attr.update(style='filled', color='white')
        for i in range(2):
            c.node(f'output_{i}', label=f'o{i}')
        c.attr(label='Output Layer')
        for i in range(6):
            for j in range(2):
                dot.edge(f'hidden3_{i}', f'output_{j}')

    return dot


neural_network = draw_neural_network()
neural_network.render('neural_network', view=True)
```
